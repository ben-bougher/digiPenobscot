```python
import numpy as np
from numpy import zeros, histogram2d as hist2d, meshgrid
from numpy.random import normal, uniform, randn, multivariate_normal
from matplotlib import pyplot as plt
import itertools
from sklearn.mixture import GMM
from scipy import linalg

from mpl_toolkits.mplot3d import axes3d

%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
def cluster(Xtest, n_clusters=7, n_iter=1000):
    # Initialize the mixture model
    clf = GMM(n_components=n_clusters,covariance_type='full', n_iter=n_iter)
    clf.fit(X)
    
    color_iter = itertools.cycle(['r', 'g', 'b', 'c','m', 'y', 'b'])
    Y_ = clf.predict(Xtest)
    splot = plt.subplot(1,1,1)
    for i, (mean, covar, color) in enumerate(zip(
                clf.means_, clf._get_covars(), color_iter)):
        v, w = linalg.eigh(covar)
        u = w[0] / linalg.norm(w[0])
        # as the DP will not use every component it has access to
        # unless it needs it, we shouldn't plot the redundant
        # components.
        if not np.any(Y_ == i):
            continue
        plt.scatter(Xtest[Y_ == i, 0], Xtest[Y_ == i, 1], .8, color=color)
       
        # Plot an ellipse to show the Gaussian component
        #angle = np.arctan(u[1] / u[0])
        #angle = 180 * angle / np.pi  # convert to degrees
        #ell = mpl.patches.Ellipse(mean, v[0], v[1], 180 + angle, color=color)
        #ell.set_clip_box(splot.bbox)
        #ell.set_alpha(0.5)
        #splot.add_artist(ell)
    plt.xlim([-1,1])
    plt.ylim([-1, 1])
    plt.show()
    
    return Y_

def IG(g,m,c,vp_av,dvp):
    I = .5*(dvp/vp_av)*(1 + g)
    G = (I/(1+g))*(1 - 4*((vp_av-c)/(vp_av*m))*((2./m)+ g*(vp_av-c)/(vp_av*m)))
        
    return I,G
        
def shueyIG(upperRock, lowerRock):
    
    vp1,vs1,rho1 = upperRock.sample()
    vp2,vs2,rho2 = lowerRock.sample()
    
    dvp = vp2-vp1
    dvs = vs2-vs1
    drho = rho2-rho1
    
    vpav = (vp1+vp2)/2.0
    vsav = (vs1+vs2)/2.0
    rhoav = (rho1 + rho2)/2.0
    
    I = (1/2.0)*((dvp/vpav) + (drho/rhoav))
    G = (1./2.)*(dvp/vpav) - 2*((vsav/vpav)**2)*((2*dvs/vsav)+(drho/rhoav))
    
    return I,G
```

```python
class Rock:
    def __init__(self,vp, vs, rho, vp_std, vs_std, rho_std):
        
        self.vp = vp
        self.vs = vs
        self.rho = rho
        self.vp_std = vp_std
        self.vs_std = vs_std
        self.rho_std = rho_std
    
    def sample(self):
        
        cov = np.array([[self.vp_std**2,.2,.2],[.2,self.vs_std**2,.2],[.2,.2,self.rho_std**2]])
        
        return multivariate_normal([self.vp, self.vs, self.rho], cov)
    
class mudLine:
    
    def __init__(self, c, g, m,vp, 
                 c_sig=0, g_sig=0, m_sig=0,
                 vp_sig=0, dvp_sig=500.0):
        
        self.c = c
        self.g = g
        self.m = m
        self.vp = vp
        
        self.c_sig = c_sig
        self.g_sig = g_sig
        self.m_sig = m_sig
        self.vp_sig = vp_sig
        self.dvp_sig = dvp_sig
    
  
    def get_vp(self):
        return self.vp + randn()*self.vp_sig
    
   
    def get_g(self):
        return self.g + randn()*self.g_sig
    
   
    def get_m(self):
        return self.m + randn()*self.m_sig
    
  
    def get_c(self):
        return self.c + randn()*self.c_sig
    
   
    def get_dvp(self):
        return uniform(-self.dvp_sig,self.dvp_sig)
    
        
    def sample(self,vp_av, dvp):
        """ 
        Returns a sampe from the distribution
        given corresponding to vp_av and dvp
        """
        
        dvp = float(dvp)
        vp_av = float(vp_av)
        g = self.get_g()
        m = self.get_m()
        c = self.get_c()
        
       
        I,G = IG(g,m,c,vp_av, dvp)
        
        return I,G
    

    def distribution(self, n):
        """
        param n: The number of distribution points to generate
        
        Returns the entire monte carlo generated distribution.
        """
        
        I = zeros(n)
        G = zeros(n)
        
        for i in range(n):
            dvp = self.get_dvp() 
            vp_av = self.get_vp()
            g = self.get_g()
            m = self.get_m()
            c = self.get_c()
        
            I[i], G[i] = IG(g,m,c,vp_av, dvp)
            
        return I,G
        
        
```

```python
ML = mudLine(1360.0, .25, 1.16, 300.,0.05,0.5)
```

```python
vp_av = 2900.0
shale = Rock(3240,1620,2340, 200.,200.,200.)
brineSand = Rock(2590,1060, 2210, 200., 200., 200.)
gasSand = Rock(2540,1620,2090, 200., 200., 200.)
shale_background = np.array([ML.sample(shale.sample()[0], i) 
                             for i in normal(0,shale.vp_std, 100)])
sand_background = np.array([ML.sample(brineSand.sample()[0], i) 
                            for i in normal(0,brineSand.vp_std, 100)])

shaleSand = np.array([shueyIG(shale, brineSand) for i in range(100)])
sandShale = np.array([shueyIG(brineSand, shale) for i in range(100)])
gasSandShale = np.array([shueyIG(shale, gasSand) for i in range(50)])
shaleGasSand = np.array([shueyIG(gasSand, shale) for i in range(50)])
```

```python
plt.scatter(shale_background[:,0], shale_background[:,1],.8, color='b')
plt.scatter(sand_background[:,0], sand_background[:,1],.8, color='b')
plt.scatter(sandShale[:,0], sandShale[:,1], .8,color='b')
plt.scatter(shaleSand[:,0], shaleSand[:,1], .8,color='b')

plt.scatter(sandShale[:,0], sandShale[:,1], .8,color='b')
plt.scatter(shaleSand[:,0], shaleSand[:,1], .8,color='b')

plt.scatter(gasSandShale[:,0], gasSandShale[:,1],.8, color='r')
plt.scatter(shaleGasSand[:,0], shaleGasSand[:,1], .8,color='r')

plt.xlim([-1,1]);plt.ylim([-1,1])
```

```python
X = np.vstack([sandShale,shaleSand, shale_background, sand_background, gasSandShale, shaleGasSand])
```

```python
clusters = cluster(X, 4, n_iter=600)
```
