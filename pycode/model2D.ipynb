```python
import requests
import h5py
from matplotlib import pyplot as plt
import numpy as np

import itertools
import noise
from scipy.io import savemat, loadmat

plt.ioff()

import os
```

```python
def shueyIG(vp, vs, rho):
    
    vp1 = vp[:-1,:]
    vp2 = vp[1:,:]
    
    vs1 = vs[:-1, :]
    vs2 = vs[1:, :]
    
    rho1 = rho[:-1, :]
    rho2 = rho[1:, :]
    
    dvp = vp2 - vp1
    dvs = vs2 - vs1
    drho = rho2 - rho1
    
    vpav = (vp1+vp2)/2.0
    vsav = (vs1+vs2)/2.0
    rhoav = (rho1 + rho2)/2.0
    
    I = (1/2.0)*((dvp/vpav) + (drho/rhoav))
    G = (1./2.)*(dvp/vpav) - 2*((vsav/vpav)**2)*((2*dvs/vsav)+(drho/rhoav))
    
    return I,G
```

```python
def get_model(name):
    
    data = requests.get('http://localhost:8080/model_data?model='+name, timeout=6000.0)
    f = h5py.File('/Users/bbougher/src/modelr/rock_model.hf5', 'r') 
    vp = f["vp"].value
    vs = f["vs"].value
    rho = f["rho"].value
    
    return vp, vs, rho
```

```python
def add_simplex(vp, vs, rho, persistence=.9, octaves=10, scaling=1500.):
    
    for x,y in itertools.product(np.arange(vp.shape[0]), np.arange(vp.shape[1])):
        
        vp[x,y] += (noise.snoise2(x/1400.,y/1400., persistence=persistence,
                                  octaves=octaves))*scaling
        vs[x,y] += (noise.snoise2(x/1400.,y/1400., persistence=persistence,
                                  octaves=octaves))*scaling
        rho[x,y] += (noise.snoise2(x/1400.,y/1400., persistence=persistence,
                                   octaves=octaves))*scaling

        
    
```

```python
def plot_model(data, filename):
    
    plt.figure()
    plt.imshow(data, cmap='gist_earth',
               clim=[1200,4000], interpolation='nearest')
    plt.colorbar()
    plt.savefig(filename)
    
    
```

```python
def plot_scatter(I,G, filename):
    
    plt.figure()
    
    plt.scatter(I.flatten()[::20], G.flatten()[::20], .8)
    plt.xlabel('Intercept')
    plt.ylabel('Gradient')
    plt.xlim([-.6,.6])
    plt.ylim([-.6,.6])

    plt.savefig(filename)
```

```python
def make_curvelets(I,G):
    
    savemat('tempIG.mat',{'I':I,'G':G})
    os.system('matlab -nojvm -nodisplay -nosplash -r "curveletTrans"')
    data = loadmat('tempCurve.mat')
    
    cI = data['cI']
    cG = data['cG']
    
    os.remove('tempIG.mat')
    os.remove('tempCurve.mat')
    
    return cI, cG
    
```

```python
def make_sparse_curvelets(I,G):
    
    savemat('tempIG.mat',{'I':I,'G':G})
    os.system('matlab -nojvm -nodisplay -nosplash -r "sparseCurveletTrans"')
    
    data = loadmat('tempCurve.mat')
    
    cI = data['cI']
    cG = data['cG']
    
    os.remove('tempIG.mat')
    os.remove('tempCurve.mat')
    
    return cI, cG
```

```python
def make_figures(model_name):
    
    vp,vs,rho = get_model(model_name)
    
    path = os.path.join('figures', model_name)
    add_simplex(vp,vs,rho)
    
    plot_model(vp, path + '_vp.png')
    
    I,G = shueyIG(vp,vs,rho)
    
    plot_scatter(I,G,path + '_scatter.eps' )
    
    cI, cG = make_curvelets(I,G)
    
    plot_scatter(cI, cG,path + '_curve_scatter.eps' )
    
    csI, csG = make_sparse_curvelets(I,G)
    
    plot_scatter(csI, csG,path + '_sparse_scatter.eps' )    
```

```python
for f in ['AvoModelling', 'Brine', 'Shale', 'GasSand']:
    make_figures(f)
```
