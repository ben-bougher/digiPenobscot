```python
import numpy as np
from matplotlib import pyplot as plt
from AvoGMM import mudLine, AvoGMM, IG, Rock, shueyIG
from sklearn.mixture import GMM
from numpy.random import normal

from scipy import stats
from scipy.io import loadmat, savemat
import warnings
%matplotlib inline
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
def scatter_plot(X, labels, reverse=False):
    
    unique_labels = np.unique(labels)
    if reverse:
        unique_labels  = unique_labels[::-1]
    
    colors = ['k','b','r','y','g','m','c']
    for label, color in zip(unique_labels, colors):
        
        index = labels==label
        data = X[index, :]
        
        plt.scatter(data[:,0], data[:,1], .8, color=color)
        plt.xlim(-1,1);plt.ylim(-1,1)
```

As a sanity check, try synthetic data with a trend line and known gaussian outlier clusters

```python
ML = mudLine(c = 1360.0, g = .25, m=1.16, vp=3000, 
             c_sig=200., g_sig=0.05, m_sig=.116, 
             vp_sig=2000.0, dvp_sig=2500.0, n=1000000)

avo = AvoGMM(ML, n_gaussians=1, n_iter=100, n_init=10,
             tol=1e-3,min_covar=1e-3, verbose=0, bias=[1,2000])
gmm = GMM(n_components=2, n_iter=100, n_init=10, min_covar=1e-3,
          covariance_type='full')

vp_av = 2900.0
shale = Rock(3240,1620,2340, 200.,200.,200.)
brineSand = Rock(2590,1060, 2210, 200., 200., 200.)
gasSand = Rock(2540,1620,2090, 200., 200., 200.)
shale_background = np.array([ML.sample(shale.sample()[0], i) 
                             for i in normal(0,shale.vp_std, 1000)])
sand_background = np.array([ML.sample(brineSand.sample()[0], i) 
                            for i in normal(0,brineSand.vp_std, 1000)])

shaleSand = np.array([shueyIG(shale, brineSand) for i in range(500)])
sandShale = np.array([shueyIG(brineSand, shale) for i in range(500)])
gasSandShale = np.array([shueyIG(shale, gasSand) for i in range(100)])
shaleGasSand = np.array([shueyIG(gasSand, shale) for i in range(100)])

X = np.vstack([sandShale,shaleSand, shale_background, 
               sand_background, gasSandShale, shaleGasSand])
true_labels = np.concatenate((np.zeros(3000), np.ones(200)))
```

# Scatter plot and the true labels

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], .8, color='k')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.savefig('../figures/synthetic_crossplot.eps')

plt.figure()
scatter_plot(X, true_labels)
plt.savefig('../figures/true_crossplot.eps')
```

## Training using avo mixing and strictly gaussian mixing

```python
avo_labels = avo.fit_predict(X)
gmm.fit(X)
gmm_labels = gmm.predict(X)
```

```python
plt.figure()
scatter_plot(X, avo_labels, reverse=True)
success = max(np.sum(avo_labels==true_labels), np.sum(avo_labels!=true_labels))
plt.title(str(success/float(true_labels.size)))
plt.savefig('../figures/MOE_crossplot.eps')
plt.figure()
success = max(np.sum(gmm_labels==true_labels), np.sum(gmm_labels!=true_labels))
scatter_plot(X, gmm_labels)
plt.title(str(success/float(true_labels.size)))
plt.savefig('../figures/GMM_crossplot.eps')

slope, intercept, r_value, p_value, std_err = stats.linregress(X[:,0],X[:,1])
plt.figure()
plt.scatter(X[:,0],X[:,1], .8, color='k')
plt.xlim(-1,1);plt.ylim(-1,1)
plt.plot(np.linspace(-1,1,100),np.linspace(-1,1,100)*slope +intercept)
```

Avo outperformed gaussian clusters as it keeps the shape of the rotated mudline curve

## Testing on slice of Penobscot data

```python
data = loadmat('../penobscot_IG/inline_1200.mat')
X = np.array([data['intData'].flatten(), 
                  np.array(data['gradData'].flatten())]).T
X/=2000.0 # normalize
```

```python
ML = mudLine(c = 1360.0, g = .25, m=1.16, vp=3000, 
             c_sig=200., g_sig=0.05, m_sig=.116, 
             vp_sig=2000.0, dvp_sig=2500.0, n=1000000)

avo = AvoGMM(ML, n_gaussians=4, n_iter=100, n_init=1,
             tol=1e-3,min_covar=1e-3, verbose=False, bias=[1,1,1,1,2000])
```

```python
labels = avo.fit_predict(X)
```

```python
scatter_plot(X, labels)
plt.savefig('PenobscotMOE.eps')
```

```python
intData = data["intData"]
gradData = data["gradData"]
clusters = np.reshape(labels, intData.shape)
```

```python
def summarize_clusters(labels, intData, gradData, file_prefix=None):
    for c in np.unique(labels):
        filteredInt = np.zeros(intData.shape)
        filteredInt[clusters==c] = intData[clusters==c]

        filteredGrad = np.zeros(gradData.shape)
        filteredGrad[clusters==c] = gradData[clusters==c]

        plt.figure()
        plt.subplot(121)
        plt.imshow(filteredInt/2000.0, clim=[-1,1],cmap='gray')
        plt.subplot(122)
        plt.scatter(filteredInt.flatten()/2000.,filteredGrad.flatten()/2000., .8, color='k')
        plt.xlim(-1,1);plt.ylim(-1,1)
        
        if file_prefix is not None:
            plt.savefig(file_prefix + str(c) +'.eps')
```

```python
summarize_clusters(clusters, intData, gradData, file_prefix='../figures/MOE')
```

Looks like a channel, a zero-zero cluster, a solid background trend, and two wierd unexplained clusters. Check to see if maybe they are side-lobes?

```python
c=np.unique(labels)[-1]
filteredInt = np.zeros(intData.shape)
addInt = np.zeros(intData.shape)
filteredInt[clusters==c] = intData[clusters==c]
plt.figure()

plt.plot(filteredInt[200:300,200], 'k')
addInt += filteredInt

c=np.unique(labels)[0]
filteredInt = np.zeros(intData.shape)
filteredInt[clusters==c] = intData[clusters==c]
#plt.figure()

plt.plot(filteredInt[200:300,200], 'b')

addInt += filteredInt
plt.figure()
plt.imshow(addInt, cmap='gray')
```

Looks like it is just filling gaps in the wavelet, so not a necessarily a side lobe issue. Look at traces and see how well peaks/troughs in Intercept and Gradient line up.

```python
plt.plot(intData[200:400,200], label='I')
plt.plot(gradData[200:400,200],'k', label='G')
plt.legend()
plt.savefig('../figures/correlation_test.eps')
```

There is significant uncorrelated patterns going that would cause random outliers in cross-plots. As a first attempt, do peak finding and only keep points where there is a peak/trough in both Intercept and Gradient.

```python
from scipy.signal import find_peaks_cwt
```

```python
peaks = []
for i in range(intData.shape[1]):
    peak_indI = find_peaks_cwt(intData[:,i], np.arange(1,2))
    trough_indI = find_peaks_cwt(-intData[:,i], np.arange(1,2))
    indI = np.concatenate((peak_indI, trough_indI))
    
    peak_indG = find_peaks_cwt(gradData[:,i], np.arange(1,2))
    trough_indG = find_peaks_cwt(-gradData[:,i], np.arange(1,2))
    indG = np.concatenate((peak_indG, trough_indG))
    
    peaks.append(np.intersect1d(indG, indI))
```

```python
intPeak = np.zeros(intData.shape)
gradPeak = np.zeros(gradData.shape)
for trace, index in enumerate(peaks):
    intPeak[index, trace] = intData[index, trace]
    gradPeak[index, trace] = gradData[index, trace]
    
```

```python
plt.scatter(intPeak.flatten()/2000., gradPeak.flatten()/2000., .8, 'k')
plt.xlim(-1,1);plt.ylim(-1,1)
plt.savefig('../figures/peak_picked.eps')
```

Looks to have removed the weirdness. Try the clustering again.

```python
avo = AvoGMM(ML, n_gaussians=3, n_iter=100, n_init=1,
             tol=1e-3,min_covar=1e-3, verbose=0, bias=[1,1,1,2000])
X = np.array([intPeak.flatten(), 
              gradPeak.flatten()]).T
X/=2000.0 # normalize

# Add a small amount of random noise so we don't have singular variance matrix for the zeros
X+=np.random.randn(X.shape[0],X.shape[1])*.00001
```

```python
labels = avo.fit_predict(X)
```

```python
clusters = np.reshape(labels, intData.shape)
scatter_plot(X, labels)
summarize_clusters(labels, intPeak, gradPeak)
```

Did what was expected, but we had to throw away a ton of data. This is because of poor processing where the intercept and gradient images weren't very well correlated. Using a joint sparse curvelet transform during the inversion will help.
